```python
import matplotlib.pyplot as plt
%matplotlib inline
from random import choice
```

```python
#reference
x=0
y=0
# eq1
x=0.5*x+0.5
y=0.5*y+0.5
# eq2
x=0.5*x+1
y=0.5*y
# eq3
x=0.5*x
y=0.5*y
# list for choice of eqn.
listy=[1,2,3]
```

```python
xa=[]
ya=[]
in_val=int(input('Enter a number of iterations '))
for i in range(in_val):
    
    ch=choice(listy)
    if ch==1:
        x=0.5*x+0.5
        y=0.5*y+0.5
    elif ch==2:
        x=0.5*x+1
        y=0.5*y
    else:
        x=0.5*x
        y=0.5*y
    xa.append(x)
    ya.append(y)
print(xa,ya)
    
        
        
           
           
        
           
    
```

```python
plt.rc('figure',figsize=(20,15))
plt.plot(xa,ya,'o')
plt.show()
# plt.savefig('SierpinskyTriangle.png')
```
