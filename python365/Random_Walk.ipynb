```python
def rand_walk(total_steps):
    step_options=[1,-1]
    step_choice=choice(step_options)
    walk=[]
    walk.append(step_choice)
    for i in range (1,total_steps):
#         next_step=choice(step_options)
        walk.append(walk[i-1]+choice(step_options))
    return walk
    
```

```python
def build_plot(walk):
    plt.rc('figure',figsize=(12,8))
    plt.plot(walk,'ko--')
    plt.show()
```

```python
walk=rand_walk(100)
```

```python
print(walk)
```

```python
build_plot(walk)
```

# functions for plotting different random walks in a plot.

```python
def rand_walks(num,steps):
    list_num=list(range(1,num+1))
    for i in range(0,num):
        x=list(range(1,steps+1))
        plt.plot(x,rand_walk(steps),label='Plot No.'+' '+str(list_num[i]))
        plt.xlabel('steps taken')
        plt.ylabel('distance from origin')
        plt.legend()
```

```python
rand_walks(6,1000)
```
